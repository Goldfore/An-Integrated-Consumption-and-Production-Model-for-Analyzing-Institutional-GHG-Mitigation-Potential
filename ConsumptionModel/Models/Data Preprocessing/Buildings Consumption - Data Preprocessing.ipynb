## Data Preprocessing

Get rid of unnecessary rows

```python
direct_consumption_data.drop(0, axis = 0, inplace = True)
direct_consumption_data.reset_index(drop=True, inplace=True)
```

```python
appliances_df = appliances_df[appliances_df['building_number'] != '67']
```

Split consumption data to buildings and Energy Center

```python
energy_center_data = direct_consumption_data.iloc[48:,:]
direct_consumption_data.drop(energy_center_data.index, axis=0, inplace=True)
energy_center_data.reset_index(drop=True, inplace=True)
```

Adjusting titles

```python
direct_consumption_data.rename(columns = {'Unnamed: 0':'building_number'}, inplace = True)

tr_consumption_data.rename(columns = {'T.R':'building_number'}, inplace = True)

energy_center_data.rename(columns = {'Unnamed: 0':'building_number'}, inplace = True)

buildings_types.rename(columns = {'Buildings':'building_number'}, inplace = True)
buildings_types.rename(columns = {'Type':'type'}, inplace = True)


buildings_details.rename(columns = {'Building':'building_number'}, inplace = True)

chemical_vaporizers.rename(columns = {'Building':'building_number'}, inplace = True)
```

Changing the type of Datasets values to numerics

```python
direct_consumption_data = direct_consumption_data.replace({',':''}, regex=True)
direct_consumption_data.iloc[:,1:] = direct_consumption_data.iloc[:,1:].apply(pd.to_numeric)

tr_consumption_data = tr_consumption_data.replace({',':''}, regex=True)
tr_consumption_data.iloc[:,1:] = tr_consumption_data.iloc[:,1:].apply(pd.to_numeric)

energy_center_data = energy_center_data.replace({',':''}, regex=True)
energy_center_data.iloc[:,1:] = energy_center_data.iloc[:,1:].apply(pd.to_numeric)
```

Remove unwanted rows & columns (Building 43 & Years 2015-2018)

```python
direct_consumption_data = direct_consumption_data[direct_consumption_data['building_number'] != 'Building 43']
direct_consumption_data.drop(direct_consumption_data.iloc[:, 1:49], axis=1, inplace = True)

tr_consumption_data = tr_consumption_data[tr_consumption_data['building_number'] != 'Building 43']
tr_consumption_data.drop(tr_consumption_data.iloc[:, 1:13], axis=1, inplace = True)

energy_center_data.drop(energy_center_data.iloc[:, 1:49], axis=1, inplace = True)

buildings_types = buildings_types[buildings_types['building_number'] != 'Building 43']

buildings_details = buildings_details[buildings_details['building_number'] != 'Building 43']

chemical_vaporizers = chemical_vaporizers[chemical_vaporizers['building_number'] != '43']

appliances_df = appliances_df[appliances_df['building_number'] != '43']
```

Sets up two groups, Labs, and Other Buildings

```python
labels_df = buildings_types.copy(deep=True)
labels_df['label'] = 0
labels_df['label'][labels_df['type'] == 'Labs'] = 1
labels_df.drop(['type'], axis=1, inplace=True)
```

Separation between Cooling and Heating

```python
heating_data = energy_center_data.drop([2,3,4,5,6], axis=0, inplace=False).reset_index(drop=True, inplace=False)
cooling_data = energy_center_data.drop([1,7], axis=0, inplace=False).reset_index(drop=True, inplace=False)
```

```python
temp_67_building_heating = heating_data.iloc[0:1,:].copy(deep=True)

temp_67_building_heating.iloc[0:1,1:] = 0

temp_67_building_heating.iloc[0:1,1:4] = heating_data.iloc[0:1,1:4]*0.7
temp_67_building_heating.iloc[0:1,4:5] = heating_data.iloc[0:1,4:5]*0.3
temp_67_building_heating.iloc[0:1,12:16] = heating_data.iloc[0:1,12:16]*0.7
temp_67_building_heating.iloc[0:1,16:17] = heating_data.iloc[0:1,16:17]*0.3
temp_67_building_heating.iloc[0:1,24:25] = heating_data.iloc[0:1,24:25]*0.7

heating_data.iloc[0:1,:] = temp_67_building_heating
heating_data
```

```python
heating_df = heating_data.iloc[:,1:].sum(axis=0).to_frame().T
```

```python
cooling_data.iloc[0:1,1:4] = cooling_data.iloc[0:1,1:4]*0.3
cooling_data.iloc[0:1,4:5] = cooling_data.iloc[0:1,4:5]*0.7
cooling_data.iloc[0:1,12:16] = cooling_data.iloc[0:1,12:16]*0.3
cooling_data.iloc[0:1,16:17] = cooling_data.iloc[0:1,16:17]*0.7
cooling_data.iloc[0:1,24:25] = cooling_data.iloc[0:1,24:25]*0.3

cooling_data
```

Combine Chemical Vaporizers with Electricity Products

```python
appliances_df = pd.merge(appliances_df, chemical_vaporizers, how='outer', on='building_number')
month_number_to_move = appliances_df.pop("month_number")
appliances_df.insert(0, "month_number", month_number_to_move)
```

Convert T.R Consumption to Indirect Consumption according to proportion

```python
indirect_consumption_data = tr_consumption_data.copy(deep=True)

indirect_consumption_data.iloc[:,1:] = indirect_consumption_data.iloc[:,1:].div(indirect_consumption_data.iloc[:,1:].sum(axis=0), axis=1)
indirect_consumption_data.iloc[:,1:] = indirect_consumption_data.iloc[:,1:].multiply(cooling_data.iloc[:,1:].sum(axis=0), axis=1)
```

Throwing the rows of the medical buildings

```python
indirect_consumption_data = indirect_consumption_data[indirect_consumption_data['building_number'] != 'Building M5-M7']
indirect_consumption_data = indirect_consumption_data[indirect_consumption_data['building_number'] != 'Building M8']
indirect_consumption_data = indirect_consumption_data[indirect_consumption_data['building_number'] != 'Building M9']
indirect_consumption_data = indirect_consumption_data[indirect_consumption_data['building_number'] != 'Building M10']
```

Adjusting Fuel Orders

```python
fuel_orders_df = pd.DataFrame(np.array([[0, 0, 0, 0, 0]]), columns=['12/18','01/19','02/19','03/19','04/19'])

for i in range(len(fuel_orders)):
    if "12/2018" in fuel_orders['Date'].iloc[i]: 
        fuel_orders_df['12/18'].iloc[0] = fuel_orders_df['12/18'].iloc[0] + fuel_orders['Quantity'].iloc[i]
    if "01/2019" in fuel_orders['Date'].iloc[i]: 
        fuel_orders_df['01/19'].iloc[0] = fuel_orders_df['01/19'].iloc[0] + fuel_orders['Quantity'].iloc[i]
    if "02/2019" in fuel_orders['Date'].iloc[i]: 
        fuel_orders_df['02/19'].iloc[0] = fuel_orders_df['02/19'].iloc[0] + fuel_orders['Quantity'].iloc[i]
    if "03/2019" in fuel_orders['Date'].iloc[i]: 
        fuel_orders_df['03/19'].iloc[0] = fuel_orders_df['03/19'].iloc[0] + fuel_orders['Quantity'].iloc[i]
    if "04/2019" in fuel_orders['Date'].iloc[i]: 
        fuel_orders_df['04/19'].iloc[0] = fuel_orders_df['04/19'].iloc[0] + fuel_orders['Quantity'].iloc[i]
        
fuel_orders_df
```

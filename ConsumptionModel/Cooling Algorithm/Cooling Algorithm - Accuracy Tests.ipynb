## Cooling Algorithm

```python
def cooling_algorithm (initialized_table):
    
    flag = True

    while (flag == True) :

        for j in range(0, initialized_table.shape[1]):
            
            for i in range(0, initialized_table.shape[0]):
                
                previous_month_value = 0
                next_month_value = 0
                previous_month_avg_trend = 0
                next_month_avg_trend = 0

                if initialized_table.iloc[i][j] == -1:
                    
                    previous_month_trend_list = []
                    next_month_trend_list = []
                    
                    if (j > 0):
                        if (initialized_table.iloc[i][j-1] != -1) and (initialized_table.iloc[i][j-1] != 0):

                            previous_month_value = initialized_table.iloc[i][j-1]

                            for k in range(0, initialized_table.shape[0]):
                                if ((initialized_table.iloc[k][j] != -1) and (initialized_table.iloc[k][j] != 0) and
                                    (initialized_table.iloc[k][j-1] != -1) and (initialized_table.iloc[k][j-1] != 0)):

                                    previous_month_trend_list.append(initialized_table.iloc[k][j] / initialized_table.iloc[k][j-1])
                                    
                            previous_month_upper_threshold = statistics.mean(previous_month_trend_list) + 2.5 * statistics.stdev(previous_month_trend_list)
                            previous_month_lower_threshold = statistics.mean(previous_month_trend_list) - 2.5 * statistics.stdev(previous_month_trend_list)
                            previous_month_trend_list = [i for i in previous_month_trend_list if (i < previous_month_upper_threshold) and (i > previous_month_lower_threshold)]
                            previous_month_avg_trend = statistics.mean(previous_month_trend_list)
                            
                    if (j < initialized_table.shape[1]-1):
                        if (initialized_table.iloc[i][j+1] != -1) and (initialized_table.iloc[i][j+1] != 0):

                            next_month_value = initialized_table.iloc[i][j+1]

                            for k in range(0, initialized_table.shape[0]):
                                if ((initialized_table.iloc[k][j] != -1) and (initialized_table.iloc[k][j] != 0) and
                                    (initialized_table.iloc[k][j+1] != -1) and (initialized_table.iloc[k][j+1] != 0)):

                                    next_month_trend_list.append(initialized_table.iloc[k][j] / initialized_table.iloc[k][j+1])

                            next_month_upper_threshold = statistics.mean(next_month_trend_list) + 2.5 * statistics.stdev(next_month_trend_list)
                            next_month_lower_threshold = statistics.mean(next_month_trend_list) - 2.5 * statistics.stdev(next_month_trend_list)
                            next_month_trend_list = [i for i in next_month_trend_list if (i < next_month_upper_threshold) and (i > next_month_lower_threshold)]
                            next_month_avg_trend = statistics.mean(next_month_trend_list)

                    if ((previous_month_value != 0) and (previous_month_avg_trend != 0) and
                        (next_month_value != 0) and (next_month_avg_trend != 0)):
                        initialized_table.at[i, initialized_table.columns[j]] = ((previous_month_value * previous_month_avg_trend)
                                                                                 + (next_month_value * next_month_avg_trend)) / 2
                    else:
                        if ((previous_month_value != 0) and (previous_month_avg_trend != 0)):
                            initialized_table.at[i, initialized_table.columns[j]] = previous_month_value * previous_month_avg_trend
                        if ((next_month_value != 0) and (next_month_avg_trend != 0)):
                            initialized_table.at[i, initialized_table.columns[j]] = next_month_value * next_month_avg_trend
                            
                    if ((previous_month_value == 0) and (next_month_value == 0)):
                        no_trend_list = []
                        if j>11:
                            no_trend_list.append(initialized_table.iloc[i][j-12])
                        if j<60:
                            no_trend_list.append(initialized_table.iloc[i][j+12])
                        initialized_table.at[i, initialized_table.columns[j]] = statistics.mean(no_trend_list)

        flag = initialized_table.isin([-1]).any().any()
        
    return initialized_table
```

```python
def accuracy_test (test_table):
    
    actual_values_dict = {}

    while len(actual_values_dict) != 10:
        row = random.randint(1,test_table.shape[0]-2)
        col = random.randint(1,test_table.shape[1]-2)
        if (test_table.iloc[row,col] != 0) and (test_table.iloc[row,col] != -1):
            actual_values_dict[(row,col)] = test_table.iloc[row,col]
            test_table.at[row, test_table.columns[col]] = -1
    
    test_table = cooling_algorithm(test_table)
    
    errors_list = []
    
    for i in actual_values_dict:
        # print ('Actual:', actual_values_dict.get(i))
        # print ('Prediction:', test_table.iloc[i[0],i[1]])
        error = abs((actual_values_dict.get(i)-test_table.iloc[i[0],i[1]])/actual_values_dict.get(i))
        # print ('Error:', error)
        errors_list.append(error)
        
    MdAPE = statistics.median(errors_list)
    # print ('MdAPE:', MdAPE)
    return MdAPE
```

```python
MdAPE_list = []

for i in range(1000):
    MdAPE_list.append(accuracy_test(tr_consumption))

print('Average MdAPE:', statistics.mean(MdAPE_list))
```
